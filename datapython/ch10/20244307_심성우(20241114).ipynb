### 5) 상자 차트(box chart)차트에서 표시되는 값들은 어떠한 의미를 - 중앙값 (Median)
상자 안에 있는 선으로 표시되며, 데이터의 중간값을 의미합니다. 데이터가 작은 순서부터 큰 순서로 정렬될 때 중앙에 위치한 값을 가리킵니다.

- 제1사분위수 (Q1)와 제3사분위수 (Q3)

**제1사분위수 (Q1)**는 하위 25% 지점의 값을 의미하며, 상자의 아래 경계선으로 표시됩니다.
**제3사분위수 (Q3)**는 상위 75% 지점의 값을 의미하며, 상자의 위 경계선으로 표시됩니다.
**상자(Box)**는 이 두 사분위수 사이의 구간을 나타내며, 이를 **사분위 범위(IQR, Interquartile Range)**라고 부릅니다. IQR은 Q3 - Q1로 계산됩니다.
- 수염 (Whiskers)
상자에서 위아래로 뻗은 선을 의미하며, 일반적으로 수염의 길이는 Q1에서 1.5 * IQR 아래쪽과 Q3에서 1.5 * IQR 위쪽 사이의 데이터 범위를 나타냅니다. 이 범위를 벗어난 값들은 이상치로 간주됩니다.

- 이상치 (Outliers)
수염 바깥에 위치한 점들로 표시되며, 일반적으로 극단적으로 높은 값이나 낮은 값을 의미합니다. 데이터의 분포와 중앙에서 많이 떨어진 값들이며, 이 값들은 별도의 점으로 차트에 표시됩니다.지 찾아보자.

```python
import matplotlib.pyplot as plt
import numpy as np

x= np.linspace(-2, 2, 11)
y1 = 2 + 2 * x
y2 = 1 + 2 * x + x ** 2

plt.plot(x,y1,marker = 'o', color ='r', linestyle = 'dashed', label ='linerar')
plt.plot(x ,y1 ,'k^:',label = 'quad' )
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('equation')
plt.show()
```

```python
plt.scatter(x,y1,marker='^')
plt.scatter(x,y2,marker= '*')
plt.show()
```

```python
month = ['7','8','9','10','11','12']
users = [456, 492, 578, 599 ,670, 854]

import matplotlib.pyplot as plt

plt.figure(figsize=(3,3)) # 그래프 크기 변경
plt.xlabel('month')
plt.ylabel('users')
plt.plot(month, users)
plt.show()
```

```python

fig, ax = plt.subplots(1,4,figsize=(10,3))
ax[0].bar(month,users)
ax[0].set_title('Bar chart')
ax[0].set_xlabel('month')

ax[1].plot(month,users)
ax[1].set_title("Line chart")

ax[2].scatter(month,users, marker ='^')
ax[2].set_title('Scatter chart')

ax[3].barh(month,users)
ax[3].set_title('Horizontal Bar')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
theta = np.linspace(0, 2 * np.pi * 6, 1000)
theta
cos = 20 * np.cos(theta)
sin = theta * np.sin(theta)

plt.figure(figsize=(10,3))
plt.plot(theta, cos, 'b--', label = 'cos')
plt.plot(theta,sin, 'r-', label ='sin' )
plt.legend()
plt.savefig('sin-cos.png', dpi=150)
plt.show()
```

```python
dach = {'lenth' : [77,78,85,83,73,77,73,80],
        'hight' : [25,28,29,30,21,22,17,35]}

samo = {'lenth' : [75,77,86,86,79,83,83,88],
       'hight' : [56,57,50,53,60,53,49,61]}
malt = {'lenth' : [34,38,38,41,30,37,41,35],
       'hight' : [22,25,19,30,21,24,28,18]}

fig, dx = plt.subplots(1, 3, figsize=(8, 3))

dx[0].scatter(dach['lenth'], dach['hight'], color='r', label='dach')
dx[0].set_title('dach')
dx[0].set_xlabel('lenth')
dx[0].set_ylabel('height')

dx[1].scatter(samo['lenth'], samo['hight'], color='b', marker='s', label='samo')
dx[1].set_title('samo')
dx[1].set_xlabel('lenth')
dx[1].set_ylabel('height')

dx[2].scatter(malt['lenth'], malt['hight'], color='g', marker='^', label='malt')
dx[2].set_title('malt')
dx[2].set_xlabel('lenth')
dx[2].set_ylabel('height')

plt.tight_layout()
plt.show()
```

```python
plt.scatter(dach['lenth'], dach['hight'], color='r', label='dach')  # 빨간색(r)으로 설정
plt.scatter(samo['lenth'], samo['hight'], color='b', marker='s', label='samo')  # 파란색(b)으로 설정
plt.scatter(malt['lenth'], malt['hight'], color='g', marker='^', label='malt')  # 초록색(g)으로 설정
plt.legend()
plt.xlabel('lenth')
plt.ylabel('hight')
plt.title('Dog size')
plt.show()
```
